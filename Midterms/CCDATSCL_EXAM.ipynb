## Variable Definitions

* **hotel**
  - H1 = Resort Hotel
  - H2 = City Hotel

* **is_canceled** - Value indicating if the booking was canceled.
  - (1) canceled
  - (0) not canceled

* **lead_time** - Number of days that elapsed between the entering date of the booking into the PMS and the arrival date

* **arrival_date_year** - Year of arrival date

* **arrival_date_month** - Month of arrival date

* **arrival_date_week_number** - Week number of year for arrival date

* **arrival_date_day_of_month** - Day of arrival date

* **stays_in_weekend_nights** - Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel

* **stays_in_week_nights** - Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel

* **adults** - Number of adults

* **children** - Number of children

* **babies** - Number of babies

* **meal** - Type of meal booked. Categories are presented in standard hospitality meal packages:
  - Undefined/SC – no meal package.
  - BB – Bed & Breakfast.
  - HB – Half board (breakfast and one other meal – usually dinner).
  - FB – Full board (breakfast, lunch and dinner)

* **country** - Country of origin. Categories are represented in the ISO 3155–3:2013 format

* **market_segment** - Market segment designation. In categories, the term “TA” means “Travel Agents” and “TO” means “Tour Operators”

* **distribution_channel** - Booking distribution channel. The term “TA” means “Travel Agents” and “TO” means “Tour Operators”

* **is_repeated_guest** - Value indicating if the booking name was from a repeated guest.
  - 1 is a repeated guest.
  - 0 is not a repeated guest.

* **previous_cancellations** - Number of previous bookings that were cancelled by the customer prior to the current booking

* **previous_bookings_not_canceled** - Number of previous bookings not cancelled by the customer prior to the current booking

* **reserved_room_type** - Code of room type reserved. Code is presented instead of designation for anonymity reasons.

* **assigned_room_type** - Code for the type of room assigned to the booking. Sometimes the assigned room type differs from the reserved room type due to hotel operation reasons (e.g. overbooking) or by customer request. Code is presented instead of designation for anonymity reasons.

* **booking_changes** - Number of changes/amendments made to the booking from the moment the booking was entered on the PMS until the moment of check-in or cancellation

* **deposit_type** - Indication on if the customer made a deposit to guarantee the booking. This variable can assume three categories:

  - No Deposit – no deposit was made.
  - Non Refund – a deposit was made in the value of the total stay cost.  
  - Refundable – a deposit was made with a value under the total cost of stay.

* **agent** - ID of the travel agency that made the booking

* **company** - ID of the company/entity that made the booking or responsible for paying the booking. ID is presented instead of designation for anonymity reasons

* **days_in_waiting_list** - Number of days the booking was in the waiting list before it was confirmed to the customer

* **customer_type** - Type of booking, assuming one of four categories:   

  - Contract - when the booking has an allotment or other type of contract associated to it.
  - Group – when the booking is associated to a group.
  - Transient – when the booking is not part of a group or contract, and is not associated to other transient booking.
  - Transient-party – when the booking is transient, but is associated to at least other transient booking

* **adr** - Average Daily Rate as defined by dividing the sum of all lodging transactions by the total number of staying nights

* **required_car_parking_spaces** - Number of car parking spaces required by the customer

* **total_of_special_requests** - Number of special requests made by the customer (e.g. twin bed or high floor)

* **reservation_status** - Reservation last status, assuming one of three categories:
  - Canceled – booking was canceled by the customer.
  - Check-Out – customer has checked in but already departed.
  - No-Show – customer did not check-in and did inform the hotel of the reason why

* **reservation_status_date** - Date at which the last status was set. This variable can be used in conjunction with the ReservationStatus to

```python
print("[INFO]")
df.info()
```

```python
# put your answer here

import matplotlib.pyplot as plt
import seaborn as sns

# ---

leadTime = df['lead_time']
adr = df['adr']
leadSkewness = leadTime.skew()
adrSkewness = adr.skew()

plt.figure(figsize=(10, 6))
sns.histplot(leadTime, kde=True, bins=50, color='blue')

plt.axvline(leadTime.mean(), color='red', linestyle='--', label=f'Mean: {leadTime.mean():.2f}')

plt.xlabel('Lead Time (days)')
plt.ylabel('Frequency')
plt.title(f'Distribution of Lead Time (Skewness: {leadSkewness:.2f})')
plt.legend()
plt.show()

# ---

plt.figure(figsize=(10, 6))
sns.histplot(adr, kde=True, bins=50, color='blue')

plt.axvline(adr.mean(), color='red', linestyle='--', label=f'Mean: {adr.mean():.2f}')
plt.xlabel('Average Daily Rate (ADR)')
plt.ylabel('Frequency')
plt.title(f'Distribution of Average Daily Rate (Skewness: {adrSkewness:.2f})')
plt.legend()
plt.show()
```

### [ANSWER]

Both the `lead_time` and `adr` are skewed positively, having significant longer right tail as seen on both plots above.

This was further proven by the peak of the curve being on the left side of the mean. Furthermore, just by doing the `.skew()` function, you could clearly see that the returned values are showing high positive numbers instead of something less than `1`.

## 4. Customer Behavior `(20 points)`

**Question:** Do repeated guests behave differently from first-time guests?

Compare repeated guests and non-repeated guests in terms of:
- Cancellation rate
- Booking behavior (e.g., booking changes or lead time)
- Use summary statistics and/or visualizations to support your answer.


```python
# put your answer here
repeater = df[df['is_repeated_guest'] == 1]
nonRepeater = df[df['is_repeated_guest'] == 0]

repeaterRate = repeater['is_canceled'].mean() * 100
nonRepeaterRate = nonRepeater['is_canceled'].mean() * 100

print(f"Repeated Guests: {len(repeater)}\t\t|\tCancellation Rate: {repeaterRate:.2f}%")
print(f"Non-Repeated Guests: {len(nonRepeater)}\t|\tCancellation Rate: {nonRepeaterRate:.2f}%")
```

```python
plt.figure(figsize=(10, 10))
plt.bar(
    [
		f'Repeated Guests ({repeaterRate:.2f}%)',
		f'Non-Repeated Guests ({nonRepeaterRate:.2f}%)'
	],
    height=[
		repeater['is_canceled'].mean(),
		nonRepeater['is_canceled'].mean()
	]
)

plt.xlabel('Guest Type')
plt.ylabel('Cancellation Status')
plt.title('Cancellation Status by Guest Type')
plt.show()

# ---

plt.figure(figsize=(10, 10))
plt.boxplot([
    repeater['lead_time'],
    nonRepeater['lead_time']],
	labels=[
    	f'Repeated Guests ({len(repeater)})',
    	f'Non-Repeated Guests ({len(nonRepeater)})'
	]
)

plt.xlabel('Guest Type')
plt.ylabel('Lead Time (Days)')
plt.title('Lead Time by Guest Type')
plt.show()
```

```python
rlen = len(repeater)
nrlen = len(nonRepeater)
guestLen = rlen + nrlen

repeaterPct = (rlen / guestLen) * 100
nonRepeaterPct = (nrlen / guestLen) * 100

print(f"Repeated Guests: {repeaterPct:.2f}%")
print(f"Non-Repeated Guests: {nonRepeaterPct:.2f}%")
```

### [ANSWER]

At first glance, talking about the cancellation rates, the repeated guests has a 14.49% cancellation rate compared to non-repeated guests who has a 37.79% cancellation rate. That alone seems to show that repeated guests may have knowledge they'll be in good hands and feel comfortable with the hotel they're booking. Connecting this to the number 2 question and this question's answer, it also shows the same pattern of longer lead time resulting to higher cancellation rate compared to those that has shorter lead time.

Furthermore, going down to the lead time, repeated guests shows significantly shorter lead time of almost 0 or around 5 with its longest seemingly sitting at around 30 to 40 days. On the other hand, the non-repeated guests has a significant amount of lead days that could reach up to (max) around 475 days.

It is also worth noting that the repeated guests makes up a very small portion of the dataset at around 3.19%.

```python
data = {
    "First-Time": {
        "No Deposit": df[(df['is_repeated_guest'] == 0) & (df['deposit_type'] == 'No Deposit')],
        "Non Refund": df[(df['is_repeated_guest'] == 0) & (df['deposit_type'] == 'Non Refund')],
        "Refundable": df[(df['is_repeated_guest'] == 0) & (df['deposit_type'] == 'Refundable')]
    },
    "Repeated": {
        "No Deposit": df[(df['is_repeated_guest'] == 1) & (df['deposit_type'] == 'No Deposit')],
        "Non Refund": df[(df['is_repeated_guest'] == 1) & (df['deposit_type'] == 'Non Refund')],
        "Refundable": df[(df['is_repeated_guest'] == 1) & (df['deposit_type'] == 'Refundable')]
    }
}

ftl = len(data['First-Time']['No Deposit']) + len(data['First-Time']['Non Refund']) + len(data['First-Time']['Refundable'])
rl = len(data['Repeated']['No Deposit']) + len(data['Repeated']['Non Refund']) + len(data['Repeated']['Refundable'])

print("=== RESULTS ===")

# === FIRST TIME === #

print("\n[FIRST-TIME | NO DEPOSIT]")
print(f"Cancellation Rate: {data['First-Time']['No Deposit']['is_canceled'].mean() * 100:.2f}%")
print(f"Average Lead Time: {data['First-Time']['No Deposit']['lead_time'].mean():.2f} days")
print(f"Count: {len(data['First-Time']['No Deposit'])}")

# ---

print("\n[FIRST-TIME | NON-REFUND]")
print(f"Cancellation Rate: {data['First-Time']['Non Refund']['is_canceled'].mean() * 100:.2f}%")
print(f"Average Lead Time: {data['First-Time']['Non Refund']['lead_time'].mean():.2f} days")
print(f"Count: {len(data['First-Time']['Non Refund'])}")

# ---

print("\n[FIRST-TIME | REFUNDABLE]")
print(f"Cancellation Rate: {data['First-Time']['Refundable']['is_canceled'].mean() * 100:.2f}%")
print(f"Average Lead Time: {data['First-Time']['Refundable']['lead_time'].mean():.2f} days")
print(f"Count: {len(data['First-Time']['Refundable'])}")

# === REPEATED === #

print("\n[REPEATED | NO DEPOSIT]")
print(f"Cancellation Rate: {data['Repeated']['No Deposit']['is_canceled'].mean() * 100:.2f}%")
print(f"Average Lead Time: {data['Repeated']['No Deposit']['lead_time'].mean():.2f} days")
print(f"Count: {len(data['Repeated']['No Deposit'])}")
# ---

print("\n[REPEATED | NON-REFUND]")
print(f"Cancellation Rate: {data['Repeated']['Non Refund']['is_canceled'].mean() * 100:.2f}%")
print(f"Average Lead Time: {data['Repeated']['Non Refund']['lead_time'].mean():.2f} days")
print(f"Count: {len(data['Repeated']['Non Refund'])}")
# ---

print("\n[REPEATED | REFUNDABLE]")
print(f"Cancellation Rate: {data['Repeated']['Refundable']['is_canceled'].mean() * 100:.2f}%")
print(f"Average Lead Time: {data['Repeated']['Refundable']['lead_time'].mean():.2f} days")
print(f"Count: {len(data['Repeated']['Refundable'])}")

# === TOTALS === #

print(f"\nFirst Time Length: {ftl} ({ftl/len(df) * 100:.2f}%)")
print(f"Repeated Length: {rl} ({rl/len(df) * 100:.2f}%)")
print(f"Total number of bookings: {ftl + rl} ({(ftl + rl)/len(df) * 100:.2f}%)")
```

```python
# Visualize cancellation rates for all combinations
cancellationRates = [
    data['First-Time']['No Deposit']['is_canceled'].mean() * 100,
    data['First-Time']['Non Refund']['is_canceled'].mean() * 100,
    data['First-Time']['Refundable']['is_canceled'].mean() * 100,
    data['Repeated']['No Deposit']['is_canceled'].mean() * 100,
    data['Repeated']['Non Refund']['is_canceled'].mean() * 100,
    data['Repeated']['Refundable']['is_canceled'].mean() * 100
]

labels = [
    f'First-Time\nNo Deposit\n({len(data['First-Time']['No Deposit'])} | {cancellationRates[0]:.2f}%)',
    f'First-Time\nNon Refund\n({len(data['First-Time']['Non Refund'])} | {cancellationRates[1]:.2f}%)',
    f'First-Time\nRefundable\n({len(data['First-Time']['Refundable'])} | {cancellationRates[2]:.2f}%)',
    f'Repeated\nNo Deposit\n({len(data['Repeated']['No Deposit'])} | {cancellationRates[3]:.2f}%)',
    f'Repeated\nNon Refund\n({len(data['Repeated']['Non Refund'])} | {cancellationRates[4]:.2f}%)',
    f'Repeated\nRefundable\n({len(data['Repeated']['Refundable'])} | {cancellationRates[5]:.2f}%)'
]

ftg = len(data['First-Time']['No Deposit']) + len(data['First-Time']['Non Refund']) + len(data['First-Time']['Refundable'])
rg = len(data['Repeated']['No Deposit']) + len(data['Repeated']['Non Refund']) + len(data['Repeated']['Refundable'])

nd = len(data['First-Time']['No Deposit']) + len(data['Repeated']['No Deposit'])
nr = len(data['First-Time']['Non Refund']) + len(data['Repeated']['Non Refund'])
rf = len(data['First-Time']['Refundable']) + len(data['Repeated']['Refundable'])

plt.figure(figsize=(12, 6))
plt.bar(labels, cancellationRates, color=['blue', 'blue', 'blue', 'green', 'green', 'green'])
plt.xlabel(f'Guest Type ({ftg} : {rg}) and Deposit Type ({nd} : {nr} : {rf})')
plt.ylabel('Cancellation Rate (%)')
plt.title('Cancellation Rates by Guest Type and Deposit Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualize lead times
plt.figure(figsize=(12, 6))
plt.boxplot([
    data['First-Time']['No Deposit']['lead_time'],
	data['First-Time']['Non Refund']['lead_time'],
	data['First-Time']['Refundable']['lead_time'],
	data['Repeated']['No Deposit']['lead_time'],
	data['Repeated']['Non Refund']['lead_time'],
	data['Repeated']['Refundable']['lead_time']
], tick_labels=labels)

plt.xlabel(f'Guest Type and Deposit Type')
plt.ylabel('Lead Time (Days)')
plt.title('Lead Time Distribution by Guest Type and Deposit Type')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
